####dButils (DB utilities) command basics

#####1. Different Databricks utilities:
credentials: DatabricksCredentialUtils -> Utilities for interacting with credentials within notebooks<br>
data: DataUtils -> Utilities for understanding and interacting with datasets (EXPERIMENTAL)<br>
**fs**: DbfsUtils -> Manipulates the Databricks filesystem (DBFS) from the console<br>
jobs: JobsUtils -> Utilities for leveraging jobs features<br>
library: LibraryUtils -> Utilities for session isolated libraries<br>
meta: MetaUtils -> Methods to hook into the compiler (EXPERIMENTAL)<br>
**notebook**: NotebookUtils -> Utilities for the control flow of a notebook (EXPERIMENTAL)<br>
preview: Preview -> Utilities under preview category<br>
**secrets**: SecretUtils -> Provides utilities for leveraging secrets within notebooks<br>
**widgets**: WidgetsUtils -> Methods to create and get bound value of input widgets inside notebooks

#####2. DB utilities widgets

**combobox**(name: String, defaultValue: String, choices: Seq, label: String): void -> Creates a combobox input widget with a given name, default value and choices<br>
**dropdown**(name: String, defaultValue: String, choices: Seq, label: String): void -> Creates a dropdown input widget a with given name, default value and choices<br>
**get**(name: String): String -> Retrieves current value of an input widget<br>
**getAll**: Map -> Retrieves a mapping of all current values of the input widgets<br>
getArgument(name: String, optional: String): String -> (DEPRECATED) Equivalent to get<br>
**multiselect**(name: String, defaultValue: String, choices: Seq, label: String): void -> Creates a multiselect input widget with a given name, default value and choices<br>
**remove**(name: String): void -> Removes an input widget from the notebook<br>
**removeAll**: void -> Removes all widgets in the notebook<br>
**text**(name: String, defaultValue: String, label: String): void -> Creates a text input widget with a given name and default value

```python
#widgets help
dbutils.widgets.help()
```

###### Remove All Existing Widgets

```python
dbutils.widgets.removeAll()
```

###### Text Widget

```python
dbutils.widgets.text("First_Name","Matai","First_Name")
```

```python
vName = dbutils.widgets.get("First_Name")
print(f"Hello {vName}!")
```

###### Dropdown Widget

```python
dbutils.widgets.dropdown("Gender","Male",["Male","Female"])
vGender = dbutils.widgets.get("Gender")
print(f"{vName} is a {vGender}.")
```

###### Combobox Widget

```python
dbutils.widgets.combobox("Country","India",["USA","UK","France","India","Nepal"])
vCountry = dbutils.widgets.get("Country")
print(f"{vName} is from {vCountry}.")
```

###### Multiselect Widget

```python
dbutils.widgets.multiselect("Hobbies","XYZ",["ABC","DEF","GHI","JKL","MNO","XYZ"])
vHobbies=dbutils.widgets.get("Hobbies")
print(f"{vName} likes {vHobbies}.")
```

###### Get All Widgets

```python
vAllWidgets=dbutils.widgets.getAll()
print(vAllWidgets)
```
