#######Create Catalog/Schema/DB/Volume

```python
%sql
Create catalog if not exists we47_catalog;
create database if not exists we47_catalog.we47_db;
create volume if not exists we47_catalog.we47_db.we47_volume;
```

######There is no foler option in the Create command to create a folder under the volumes
######To create a folder use dbutil.fs.mkdirs

```python
dbutils.fs.mkdirs("/Volumes/we47_catalog/we47_db/we47_volume/folder1")
```

######Create a sample data under the folder

```python
sample_data = 'samaple data'
dbutils.fs.put("/Volumes/we47_catalog/we47_db/we47_volume/folder1/sample.txt",sample_data,True)
dbutils.fs.ls("/Volumes/we47_catalog/we47_db/we47_volume/folder1")
dbutils.fs.head("/Volumes/we47_catalog/we47_db/we47_volume/folder1/sample.txt")
```
