####Learning Notebook and magic command basics

#####Commonly Used Magic Commands<br>
Commands which starts with a % are magic commands<br>
- %md
- %run
- %sh
- %fs
- %python
- %sql

#####%md Markdown command
Use # for heading and sub-headings. No. of #'s from 1 to 6 creates indented headings.<br>
Example:<br>
#Heading1<br>
##Heading2<br>
###Heading3<br>
....<br>
######Heading6

Text formatting and markups examples using %md command

- Bullet points using a dash(-)
- **Bold using&nbsp;2&nbsp;***
- <b>Bold using \<b> tag </b>
- *Italic using 1&nbsp;**
- <i>Italic using \<i> tag </i>
- Insert URL using **\[Url Label](url)**: [URL](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQjHkPO9BNRX0bjAo1Q9J7sJWVrdFm_ddlP7Q&s)
- Insert image using **\!\[](url)** : ![](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQjHkPO9BNRX0bjAo1Q9J7sJWVrdFm_ddlP7Q&s)
- Text font color:
$${\color{green}This&nbsp;is&nbsp;Green&nbsp;Color}$$
$${\color{blue}\textbf{This is a bold and blue color text}}$$
$${\color{red}\textit{This is a red and italics text}}$$
